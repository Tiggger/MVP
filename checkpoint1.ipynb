Monte-Carlo Integrator

```python
#defining function to test Monte Carlo Integrator with
#-x^2-y^2

def func(r):
    #unpack coords
    x=r[0]
    y=r[0]

    return -x**2-y**2
```

```python
import numpy as np
from scipy import stats

def monte_carlo_integrate(func, bounds, n=10000):
    """
    Monte Carlo integration over specified bounds.
    
    Parameters:
    -----------
    func : callable
        Function to integrate
    bounds : list of tuples
        [(x_min, x_max), (y_min, y_max), ...]
    n : int
        Number of random samples
    """
    dim = len(bounds)
    
    # Generate uniform random points
    samples = np.zeros((n, dim))
    for d in range(dim):
        samples[:, d] = np.random.uniform(bounds[d][0], bounds[d][1], n)
    
    # Evaluate function at sample points
    func_values = np.apply_along_axis(func, 1, samples)
    
    # Calculate volume
    volume = 1.0
    for d in range(dim):
        volume *= (bounds[d][1] - bounds[d][0])
    
    # Estimate integral
    integral = volume * np.mean(func_values)
    
    # Estimate error (standard error of the mean)
    std_dev = np.std(func_values)
    error = volume * std_dev / np.sqrt(n)
    
    return integral, error, samples

# Example usage
def example_func(point):
    x, y = point
    return np.exp(-(x**2 + y**2))  # 2D Gaussian

# Define integration bounds
bounds = [(-2, 2), (-2, 2)]  # x and y limits

# Perform integration
integral, error, samples = monte_carlo_integrate(example_func, bounds, n=1000000)

print(f"Estimated integral: {integral:.6f}")
print(f"Estimated error: {error:.6f}")
print(f"95% confidence interval: [{integral - 2*error:.6f}, {integral + 2*error:.6f}]")

```
